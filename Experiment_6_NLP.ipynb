Experiment 6 Text processing


```python
import spacy
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple is looking at buying U.K. startup for $1 billion in 2024."
doc = nlp(text)
```

Tokenization

```python
tokens = [token.text for token in doc]
print("Tokens:", tokens)
```

Stemming

```python
stemmer = PorterStemmer()
stems = [stemmer.stem(token.text) for token in doc if token.is_alpha]
print("Stemming:", stems)
```

Lemmatization

```python
lemmas = [token.lemma_ for token in doc]
print("Lemmatization:", lemmas)
```

POS Tagging

```python
pos_tags = [(token.text, token.pos_) for token in doc]
print("POS Tags:", pos_tags)
```

Named Entity Recognition

```python
entities = [(ent.text, ent.label_) for ent in doc.ents]
print("Named Entities:", entities)
```

Stop word removal

```python
filtered_words = [token.text for token in doc if not token.is_stop and token.is_alpha]
print("Stop Words Removed:", filtered_words)
```
