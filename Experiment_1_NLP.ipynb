Tokenization = wtever input u give, it breaks down sentence into individual words. 2 tokenisations below- word and sentence

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

Give input below and check output( here it splits the sentence into words)

nlp below is the entire pipeline, output is in the form of document

```python
import spacy
nlp = spacy.load('en_core_web_sm')
input = "This is the second sentence. I am fine, how about you?"
doc = nlp(input)
```

```python
for token in doc:
  print(token)
```

splitting documents into sentence below

```python
for sent in doc.sents:
  print(sent)
```
