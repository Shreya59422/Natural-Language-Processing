```python
import re

```

```python
pattern = r'\w+@\w+\.\w+'
data['twitts'].apply(lambda x:re.findall(pattern=pattern,string = x))
```

fixing the boundary

```python
pattern = r'\b[A-Za-z._+-]+@[A-Za-z.].[A-Z|a-z]\b'
data['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
```

{} represents after . , there will be possibility to consider two characters

```python
pattern = r'\b[A-Za-z._+-]+@[A-Za-z.].[A-Z|a-z]{2,}\b'
data['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
```

URLs: 2 cases

https://gmail.com

www.google.com


captial S represents wtever it is available upto space

```python
pattern =r'\b[A-Za-z0-9._%+-]+@[A-ZA-ZA-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' #word boundary
data['email']= data['twitts'].apply(lambda x:",".join(re.findall(pattern=pattern,string = x)))
data['email']
```

```python
pattern =r'\b[A-Za-z0-9._%+-]+@[A-ZA-ZA-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' #word boundary
data['count']= data['twitts'].apply(lambda x:x.count(',')+1 if len(x)>0 else 0)
data
```

```python
pattern =r'http\S+|www\S+'
data['URLs']= data['twitts'].apply(lambda x:",".join(re.findall(pattern=pattern,string = x)))
data['URL_count']= data['twitts'].apply(lambda x:x.count(',')+1 if len(x)>0 else 0)
data
```
